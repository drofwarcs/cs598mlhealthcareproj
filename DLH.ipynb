```python
dataset = intermediate.merge(hf_diagnoses, how="left", on="HADM_ID")
dataset
```

```python
#Drop all rows where the patient was NOT diagnosed with a heart failure related ICD9 code
dataset = dataset.dropna(subset=["ICD9_CODE"]).reset_index()
dataset.head()
```

```python
len(dataset)
```

```python
#Drop all rows where a discharge summary does not exist
dataset = dataset.dropna(subset=["CATEGORY"]).reset_index()
dataset.head()
```

```python
len(dataset)
```

```python
dataset = dataset.drop(["level_0", "index"], axis=1)
dataset.head()
```

```python
sum(dataset.duplicated())
```

```python
dataset["ADMITTIME"] = pd.to_datetime(dataset["ADMITTIME"])
dataset["DISCHTIME"] = pd.to_datetime(dataset["DISCHTIME"])
```

```python
dataset.head()
```

```python
dataset["SAME_SUBJECT_ID"] = dataset["SUBJECT_ID_x"] == dataset["SUBJECT_ID_x"].shift(1)
dataset["SAME_HADM_ID"] = dataset["HADM_ID"] == dataset["HADM_ID"].shift(1)
```

```python
dataset
```

```python
readmit_times = []

subject_id = 0
hadm_id = 0
dischtime = pd.Timedelta(0)
for i in range(len(dataset)):
    
    if (dataset["SUBJECT_ID_x"][i] == subject_id) and (dataset["HADM_ID"][i] != hadm_id):
        readmit_times.append(dataset['ADMITTIME'][i] - dischtime)
        
        if dataset["SAME_HADM_ID"][i]:
            dischtime = dataset["DISCHTIME"][i]
    elif (dataset["SUBJECT_ID_x"][i] != subject_id):
        subject_id = dataset["SUBJECT_ID_x"][i]
        hadm_id = dataset["HADM_ID"][i]
        dischtime = dataset["DISCHTIME"][i]
        
        #If not a readmission at all, value 0
        readmit_times.append(pd.Timedelta(0))
    else: 
        readmit_times.append(pd.Timedelta(0))
```

```python
readmit_times
```

```python
dataset["READMISSION_TIMES"] = readmit_times
```

```python
dataset[dataset["SUBJECT_ID_x"] == 357]
```

```python
dataset.head()
```

```python
general_readmission = []

subject_id = 0
for i in range(len(dataset)):
    if (dataset["SUBJECT_ID_x"][i] != subject_id) and (dataset["READMISSION_TIMES"].shift(-1)[i] > pd.Timedelta(0)):
        general_readmission.append(1)
        
        subject_id = dataset["SUBJECT_ID_x"][i]
    
    elif (dataset["SAME_SUBJECT_ID"][i] != dataset["SAME_SUBJECT_ID"].shift(-1)[i]) or (dataset["SAME_SUBJECT_ID"][i] != dataset["SAME_SUBJECT_ID"].shift(-2)[i]):
        general_readmission.append(0)
        
    elif (dataset["READMISSION_TIMES"][i] > pd.Timedelta(0)):
        general_readmission.append(1)
        
    else:
        general_readmission.append(0)
```

```python
general_readmission
```

```python
dataset["GENERAL_READMISSION"] = general_readmission
```

```python
dataset[dataset["SUBJECT_ID_x"] == 357]
```

```python
thirtyday_readmission = []

subject_id = 0
for i in range(len(dataset)):
    if (dataset["SUBJECT_ID_x"][i] != subject_id) and (dataset["READMISSION_TIMES"].shift(-1)[i] > pd.Timedelta(0)) and (dataset["READMISSION_TIMES"].shift(-1)[i] < pd.Timedelta(days=30)):
        thirtyday_readmission.append(1)
        
        subject_id = dataset["SUBJECT_ID_x"][i]
    
    elif (dataset["SAME_SUBJECT_ID"][i] != dataset["SAME_SUBJECT_ID"].shift(-1)[i]) or (dataset["SAME_SUBJECT_ID"][i] != dataset["SAME_SUBJECT_ID"].shift(-2)[i]):
        thirtyday_readmission.append(0)
        
    elif (dataset["READMISSION_TIMES"][i] < pd.Timedelta(days=30)) and (dataset["READMISSION_TIMES"][i] > pd.Timedelta(0)):
        thirtyday_readmission.append(1)
        
    else:
        thirtyday_readmission.append(0)
```
